```python
def make_new_col(col_name):
    tuples_list=[]
    value_list=[]
    for i in range(len(df)):
        value_list.append(df.loc[i,col_name])
    tuples_list.append((col_name,value_list))
    return dict(tuples_list)
```

```python
def make_new_df(df):
    dic0={}
    for i in range(len(df.columns)):
        dic1=make_new_col(df.columns[i])
        dic0.update(dic1)
    new_df=pd.DataFrame(dic0)
    new_df['Timestamp']=pd.date_range('01/01/2020',periods=len(df),freq='0.186T')
    return new_df
new_df=make_new_df(df)
new_df
```

```python
def isfloat(value):
    try:
        float(value)
        return True
    except ValueError:
        return False
```

```python
nan=float('nan')
```

```python
def typesafe_isnan(obj):
    return isinstance(obj, float) and obj != obj
```

```python
def find_avg(index,col_name):
    value_list=[]
    temp_list=[]
    for i in range (len(df)):
        isum=0
        times=0
        if (pd.to_datetime(df['Timestamp'][i])==(pd.to_datetime(df['Timestamp'][index])+dt.timedelta(minutes=15))):
            temp_list.append(df.at[i,col_name])
    for j in range(len(temp_list)):
        if (typesafe_isnan(temp_list[j])==False):
            isum=isum+float(temp_list[j])
            times=times+1
    if (times!=0):
        value_list.append(isum/times)
    elif (times==0):
        value_list.append(float('nan'))
    #print(temp_list)
    temp_list.clear()
    return (value_list)[0]
find_avg(0,'GeoHeader GLYHRFlow1(gal/min)')
```

```python
def make_col(starting_index,col_name):
    index_list=find_index(starting_index)
    new_df = pd.DataFrame(columns = [col_name], index=[]) 
    for i in range(len(index_list)):
        new_df.append({col_name:find_avg(i,col_name)}, ignore_index=True)
    return new_df
make_col(0,'GeoHeader GLYHRFlow1(gal/min)')
```

```python
def make_rows(index):
    dic0={'Timestamp':(pd.to_datetime(df['Timestamp'][index]))}
    col_row_tuples=[]
    for i in range(1,len(df.columns)):
        col_row_tuples.append((df.columns[i],find_avg(index,df.columns[i])))
    dic1=dict(col_row_tuples)
    dic0.update(dic1)
    return_df=new_df.append(dic0, ignore_index=True)
    return return_df
new_df=make_rows(0)
new_df=make_rows(100)
```
